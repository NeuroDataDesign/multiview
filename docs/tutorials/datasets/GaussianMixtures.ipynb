# Multiview Data from Gaussian Mixtures

```python
from mvlearn.datasets import GaussianMixture
from mvlearn.utils.plotting import plot_2views
import numpy as np

%matplotlib inline
```

```python
## Create the model
n = 100
mu = [[0,1], [0,-1]]
sigma = [np.eye(2), np.eye(2)]
class_probs = [0.5, 0.5]
GM = GaussianMixture(n,mu,sigma,class_probs=class_probs)

## Transform the data using a polynomial function and add 2 noise dimensions
GM = GM.sample_views(transform='poly', n_noise=2)
```

```python
## Plot first latent against transformed second latent
Xs,y = GM.get_Xy(latents=False) 
plot_2views(Xs, title='Latent Dimensions vs. Transformed Latent Dimensions')
```
